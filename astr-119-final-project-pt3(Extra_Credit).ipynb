```python
import numpy as np
import sep
from astropy.utils.data import get_pkg_data_filename
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib import rcParams

%matplotlib inline 

rcParams['figure.figsize'] = [5.,5.]
```

# DATA105

```python
image_data105 = get_pkg_data_filename("/Users/nikosmynhier/Desktop/hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits")
data105 = fits.getdata(image_data105, ext=0)
data105 = data105.byteswap().newbyteorder()


m, s = np.mean(data105), np.std(data105)
plt.imshow(data105, interpolation='nearest', cmap='gray', vmin=m-s, vmax=m+s, origin='lower')
plt.colorbar()
plt.savefig('data105.png')
```

```python
bkg105 = sep.Background(data105)
```

```python
print(bkg105.globalback)
print(bkg105.globalrms)
```

```python
bkg105_image = bkg105.back()
```

```python
plt.imshow(bkg105_image, interpolation='nearest', cmap='gray', origin='lower')
plt.colorbar()
plt.savefig('bkg105_image.png')
```

```python
bkg105_rms = bkg105.rms()
```

```python
plt.imshow(bkg105_rms, interpolation='nearest', cmap='gray', origin='lower')
plt.colorbar()
plt.savefig('bkg105_rms.png')
```

```python
data_sub105 = data105 - bkg105
```

```python
plt.imshow(data_sub105, interpolation='nearest', cmap='gray', vmin=m-s, vmax=m+s, origin='lower')
plt.colorbar()
plt.savefig('data105-bkg105.png')
```

```python
data_sub105_min = data_sub105.min()
data_sub105_max = data_sub105.max()

b = 41.6*(data_sub105 - data_sub105_min)/(data_sub105_max - data_sub105_min)
print(b)   #to check the intensity of blue channel
```

```python
data_sub125_min = data_sub125.min()
data_sub125_max = data_sub125.max()

g = 3449*(data_sub125 - data_sub125_min)/(data_sub125_max - data_sub125_min) 
print(g)   # to check the intensity of green channel
```

```python
data_sub160_min = data_sub160.min()
data_sub160_max = data_sub160.max()

r = 41.6*(data_sub160 - data_sub160_min)/(data_sub160_max-data_sub160_min) 
print(r)   # to check the intensity of red channel
```

```python
import matplotlib.colors as colors

rgb_image = np.zeros((3600,3600,3))

rgb_image[:,:,0] = r
rgb_image[:,:,1] = g
rgb_image[:,:,2] = b
```

```python
plt.imshow(rgb_image, interpolation='nearest', origin='lower')
plt.colorbar()
```
